```python
import re #regular expression to verify the password
```

```python
#Q1 - (A)
pattern_upper_case = r'(?=.*[A-Z])' #at least 1 upper case letter
pattern_characters = r'\w{9,}' #at least 9 characters

#enter string
test_string = input("Please enter a password: ")

#Since the return value is message or "None", we need to convert it into boolean.
flag_uc = bool(re.match(pattern_upper_case, test_string))
flag_d = bool(re.match(pattern_characters, test_string))

if flag_uc == True and flag_d == True:
  print("Password is valid.")
else:
  if flag_uc != True:
    print("Your password should be included at least 1 uppercase letter.")
  if flag_d != True:
    print("Your password should be included at least 9 characters.")
```

```python
#Q1 - (B)
pattern_upper_case = r'(?=.*[A-Z])'
pattern_lower_case = r'(?=.*[a-z])'
pattern_digit = r'\w{9,}'

#enter string
test_string = input("Please enter a password: ")

#Since the return value is message or "None", we need to convert it into boolean.
flag_uc = bool(re.match(pattern_upper_case, test_string))
flag_lc = bool(re.match(pattern_lower_case, test_string))
flag_d = bool(re.match(pattern_characters, test_string))

if flag_uc == True and flag_lc == True and flag_d == True:
  print("Password is valid.")
else:
  if flag_uc != True:
    print("Your password should be included at least 1 uppercase letter.")
  if flag_lc != True:
    print("Your password should be included at least 1 lowercase letter.")
  if flag_d != True:
    print("Your password should be included at least 9 characters.")
```
